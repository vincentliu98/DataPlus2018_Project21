```python
'''
        Author: Vincent Liu
        Description: implement collaborative filtering for students
'''
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/derekliu/Desktop/Generated_Profiles.csv')
```

```python
''' 
    Create user profile
'''

# create my profiel --- input the index of the program
my_profile = [5, 16, 17, 46, 72, 102]
my_act = [df[df.columns[my_profile[i]+1]].values[0] for i in range(0, len(my_profile))]
# initial list of size of the programs
initial = (df.shape[1]-1) * [0]
# change binary number
for i in range(0, len(my_profile)):
    initial[my_profile[i]] = 1
```

```python
''' 
    Calculate cosine distance between students
''' 
from scipy.spatial.distance import cosine

# similarity scores
scores = []
# append distances to a list
for x in range(1, df.shape[0]):
    row = df.iloc[x].values.tolist()
    row = row[1:]
    if (np.count_nonzero(list(map(int, row))) == 0) :
        scores.append(2)
        continue
    scores.append(cosine(initial, list(map(int, row))))
```

```python
'''
    Find unique small numbers, because 1 is the minority, and that's what we need to compare
'''
# number of small numbers in scores to choose from
min_num = 4
score_vals = set()
for i in range(0, len(scores)):
    score_vals.add(scores[i])
min_vals = []
counter = 0
for i in sorted(score_vals):
    if counter < min_num:
        min_vals.append(i)
        counter+=1
    else:
        break
```

```python
''' 
    Initialize a python dictionary with activities done by user's neighbors and the frequency
'''

acts = {}

for i in range(0, len(min_vals)):
    my_index = [index for index, value in enumerate(scores) if value == min_vals[i]]
    for n in range(0, len(my_index)):
        val = df.iloc[my_index[n]+1]
        val = val[1:]
        index = [i for i, e in enumerate(val) if e != '0']
        for j in range(0, len(index)):
            activity = df[df.columns[index[j]+1]].values[0]
            if activity in my_act:
                continue
            if activity not in acts:
                acts[activity] = 1
            else:
                acts[activity] += 1
```

```python
# sort the programs appearing in the result and rank them in descending order
# col_1 = [item for item in col_1 if activity not in my_act]
sorted_by_value = sorted(acts.items(), key=lambda kv: kv[1], reverse = True)
col_1 = []
col_2 = []
for i in range(len(sorted_by_value)):
    col_1.append(sorted_by_value[i][0])
    col_2.append(sorted_by_value[i][1])
d = {'Recommendations': col_1, 'Score': col_2 }
recommdations = pd.DataFrame(data=d)
print('Your input:\n')
print(my_act)
recommdations
```
